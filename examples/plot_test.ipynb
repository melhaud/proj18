```python
import sys
sys.path.append('../src')
from plotting_utils import *
```

```python
path_cats_test = '../embeddings/SimCLR/test/test_cat.csv'
path_cats_train = '../embeddings/SimCLR/train/train_cat.csv'
path_cats_test_folder = '../embeddings/SimCLR/test'
```

# Metrics Benchmarking for generated CIFAR10 embeddings.

## 1. SimCLR embeddings

###  1.1 Hausdorff metric

```python
hists_abnorm, hists_norm = get_boot_strapped_hist(
    path_to_train_norm_class=path_cats_train,
    path_to_test_norm_class=path_cats_test,
    path_to_test_classes_folder=path_cats_test_folder,
    method='dorff')
```

```python
hists_abnorm.keys()
hists_abnorm.pop('test_cat', None)
hists_norm.pop('test_cat', None)

plot_hists(hists_abnorm, hists_norm, save=False)
```

## 1.2 cdist metrics

```python
hists_abnorm, hists_norm = get_boot_strapped_hist(
    path_to_train_norm_class=path_cats_train,
    path_to_test_norm_class=path_cats_test,
    path_to_test_classes_folder=path_cats_test_folder,
    method='cdist')

hists_abnorm.keys()
hists_abnorm.pop('test_cat', None)
hists_norm.pop('test_cat', None)

plot_hists(hists_abnorm, hists_norm, save=False)
```

# Conclusion
As you could see, the cdist metric performs better than Hausdorff distance. The separation between normal and outliers images may be seen with naked eye. For, example, it allows to fully separate normal objects from airplanes, while Hausdorff fails.
